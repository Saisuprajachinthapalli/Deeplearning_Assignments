```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

columns=['Petal Length', 'Petal Width', 'Sepal Length', 'Sepal Width', 'Species Type']

iris_data = pd.read_csv('/content/sample_data/iris.csv',names=columns)

Y = iris_data.pop('Species Type')
X = iris_data

# split X, Y into X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

model = Sequential()

# hidden layer uses relu
# output layer uses sigmoid

model.add(Dense(12, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

# Compile model
# use accuracy and

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model

model.fit(X, Y, epochs=50, batch_size=10)

# check the performance of the model
# use evaluate, predict, etc

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

# 2) Change epochs, batch_size, and see the changes in performance. Try at least THREE different combinations

print(" 1st time changing epochs, batch_size")

model.fit(X_train, Y_train, batch_size=25, epochs=45)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

print(" 2nd time changing epochs, batch_size")

model.fit(X_train, Y_train, batch_size=35, epochs=55)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

print(" 3rd time changing epochs, batch_size")

model.fit(X_train, Y_train, batch_size=40, epochs=60)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

# 3) change activation function in hidden layers to {sigmoid, tanh, etc}, and compare its respective performance.

print('changing hidden layers')

model.add(Dense(12, input_dim=4, activation='sigmoid'))

model.add(Dense(8, activation='tanh'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, Y, epochs=50, batch_size=10)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

# 4) add additional 1 or 2 hidden layer and see the changes in performance.

print('adding new hidden layers')

model.add(Dense(12, input_dim=4, activation='relu'))

model.add(Dense(units=64, activation='relu'))

model.add(Dense(units=32, activation='relu'))

model.add(Dense(8, activation='tanh'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, Y, epochs=50, batch_size=10)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)

# 5) change optimizer to other optimizer (e.g., rmsprop, adadelta, etc) and see the changes in performance.

print('Changing Optimizer to rmsprop')

model = Sequential()

model.add(Dense(12, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.fit(X, Y, epochs=50, batch_size=8)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)


print('Changing Optimizer to adadelta')

model = Sequential()

model.add(Dense(12, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adadelta', metrics=['accuracy'])

model.fit(X, Y, epochs=50, batch_size=10)

model.evaluate(X_train,Y_train)

Y_predictions = model.predict(X_test)

print('prediction: ', Y_predictions.flatten())
print('Y_test: ', Y_test)








```
